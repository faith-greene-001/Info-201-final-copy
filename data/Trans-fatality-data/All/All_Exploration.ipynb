This exploratory Jupyter notebook ingests an export of data from [TDoR Info](https://tdor.translivesmatter.info/) through 2020, merges on manual remarks from [Gayta Science's yearly TDoR visualizations](https://www.gaytascience.com/tdor-dataviz/), does some light cleaning, and exports a full .csv file.

TW: death, murder, rape, suicide, genocide, weapons, violence, transphobia, sex 

```python
# Import packages
import pandas as pd

# Notebook settings
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_colwidth', None)
```

```python
# Import all data
path = "./tdor_export_73.133.212.249_2020-12-29T23_48_37/tdor_export.csv"
alldata = pd.read_csv(path)
alldata
```

```python
# Summary table of cause of death
alldata.groupby(['Cause of death']).agg({'Name': 'count'}).reset_index().sort_values(by=['Name'], ascending=False)
```

```python
# Description content
alldata['Description'].tail(2)
```

```python
# Tweet content
alldata['Tweet']
```

```python
# Import manual data
manualdata = pd.read_csv('TDoR2018-20_Manual.csv', engine='python')
manualdata = manualdata[['name', 'age', 'date', 'remarks', 'link']]
manualdata
```

```python
# Merge by link
merged = alldata.merge(manualdata, how='left', left_on='Permalink', right_on='link')
```

```python
# Drop columns used to check
merged = merged.drop(columns=['name', 'age', 'date', 'link'])
merged
```

```python
# Replace manual HTML substitutions done on manual "remarks"
merged['remarks'] = merged['remarks'].str.replace('&#233;', 'é', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#250;', 'ú', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#231;', 'ç', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#231;', 'ç', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#225;', 'á', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#227;', 'ã', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#237;', 'í', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#243;', 'ó', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#246;', 'ö', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#234;', 'ê', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#252;', 'ü', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#226;', 'â', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#244;', 'ô', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#241;', 'ñ', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#193;', 'Á', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#305;', 'ı', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#305;', 'ı', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#205', 'Í', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#224;', 'à', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#350;', 'Ş', regex=False)
merged['remarks'] = merged['remarks'].str.replace('&#287;', 'ğ', regex=False)
```

```python
# Tweet Option (remove hashtags, line breaks, gaps)
merged['Tweet_edited'] = merged.Tweet.str.replace(' #trans ', ' trans ', regex=False)
merged['Tweet_edited'] = merged.Tweet_edited.str.replace(' #TransTwitter ', ' Trans Twitter ', regex=False)
merged['Tweet_edited'] = merged.Tweet_edited.str.replace(r' #[a-zA-Z]+', ' ', regex=True)
merged['Tweet_edited'] = merged.Tweet_edited.str.replace('\n', ' ', regex=False)
merged['Tweet_edited'] = merged.Tweet_edited.str.replace('\r', ' ', regex=False)
merged['Tweet_edited'].str.strip()
merged[['Tweet', 'Tweet_edited']].tail(200)
```

```python
# Export
merged.to_csv('merged.csv', encoding='utf-8-sig')
```
